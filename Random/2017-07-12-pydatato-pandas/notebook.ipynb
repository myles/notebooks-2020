```python
idx = pd.IndexSlice

def clean_date(df):
    return pd.to_datetime(df[['year', 'month', 'day']],
                          errors='cerror')
```

```python
def standardize(x):
    return (x - x.mean() / x.std())

surveys_df[['hindfoot_length', 'weight']].apply(standardize).tail()
```
