[BrandColors](https://brandcolors.net/) is a SCSS stylesheet that proviedes colours of major brands.

```python
with open('./stylesheets/brandcolors-1505508688.scss') as fobj:
    stylesheet = fobj.read()
```

```python
import re

RE_COLOUR_VAR = re.compile(r'\$bc-(?P<brand>[-\w]+): (?P<hex>[#\w]{7});')

match = RE_COLOUR_VAR.findall(stylesheet)

match[:10]
```

```python
RE_BRAND_ALT = re.compile(r'(?P<name>[-\w]+)-\d+')

for brand, colour_hex in match:
    if not RE_BRAND_ALT.match(brand):
        print('  {0}: $brand-colour-{0},'.format(brand))
```
