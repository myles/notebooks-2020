```python
import pandas as pd
```

```python
class NAICSCode:

    def __init__(self):
        self.children = []

    def __repr__(self):
        return f"NAICSCode({self.code})"

    def get_children(self):
        return self.children
    
    @classmethod
    def load(cls, row):
        naics = cls()
        
        naics.code = row['Code']
        naics.level = row['Level']
        naics.hierarchical_structure = row['Hierarchical structure']
        naics.class_title = row['Class title']
        naics.superscript = row['Superscript']
        nasics.class_definition = row['Class definition']
        
        return nasics
```
